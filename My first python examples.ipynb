## Calculate Letter Grading 
- 2 or under D
- 3 or under C
- 3.8 or under B
- Above 3.8 A

```python
student_gpa = float (input("What is your GPA ?"))

if student_gpa <= 2:
    letter_grade = "D"
elif student_gpa <= 3:
    letter_grade = "C"
elif student_gpa <= 3.8:
    letter_grade = "B"
else: 
    letter_grade = "A"

print(f"Student's grade is {letter_grade}")
```
